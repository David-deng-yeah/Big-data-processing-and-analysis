```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

# 1GB=1024* 1024 *1024字节

```python
size_GB=1024*1024*1024
list_GB=[(1,1)]
```

```python
size_GB
```

```python
sys.getsizeof(np.random.random([10000000,14]))
```

```python
for N in range(10000000,0,-100000):  
    print('N = ',N)
    D=list_GB[-1][1]# 取上一个N的维度D
    arr1 = np.random.random([N,D]) # 元素类型为float64
    while sys.getsizeof(arr1) < size_GB:
        D=D+1
        arr1=np.random.random([N,D])
    list_GB.append((N,D))
```

```python
list_GB=list_GB[1:][::-1]
```

##  散点图

```python
y_GB=[x[1] for x in list_GB]
x_GB=[x[0] for x in list_GB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
# plt.xlim([-1,2000])
plt.scatter(x_GB,y_GB)
plt.grid()
```

## 0-1000位置的散点图

```python
y_GB=[x[1] for x in list_GB]
x_GB=[x[0] for x in list_GB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
plt.xlim([-1,1000000])
plt.scatter(x_GB,y_GB)
plt.grid()
```

##  拟合曲线图

```python
y_GB=[x[1] for x in list_GB]
x_GB=[x[0] for x in list_GB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
# plt.xlim([-1,2000])
plt.plot(x_GB,y_GB)
plt.grid()
```

## 1-1000位置的拟合曲线图

```python
y_GB=[x[1] for x in list_GB]
x_GB=[x[0] for x in list_GB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
plt.xlim([-1,1000000])
plt.plot(x_GB,y_GB)
plt.grid()
```
