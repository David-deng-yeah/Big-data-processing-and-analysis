```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

# 1MB=1024*1024字节

```python
size_MB=1048576
list_MB=[(1,1)]
```

```python
for N in range(10000,0,-10):  
    print('N = ',N)
    D=list_MB[-1][1]# 取上一个N的维度D
    arr1 = np.random.random([N,D]) # 元素类型为float64
    while sys.getsizeof(arr1) < size_MB:
        D=D+1
        arr1=np.random.random([N,D])
    list_MB.append((N,D))
```

```python
list_MB=list_MB[1:][::-1]
```

##  散点图

```python
y_MB=[x[1] for x in list_MB]
x_MB=[x[0] for x in list_MB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
# plt.xlim([-1,2000])
plt.scatter(x_MB,y_MB)
plt.grid()
```

## 0-1000位置的散点图

```python
y_MB=[x[1] for x in list_MB]
x_MB=[x[0] for x in list_MB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
plt.xlim([-1,1000])
plt.scatter(x_MB,y_MB)
plt.grid()
```

##  拟合曲线图

```python
y_MB=[x[1] for x in list_MB]
x_MB=[x[0] for x in list_MB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
# plt.xlim([-1,2000])
plt.plot(x_MB,y_MB)
plt.grid()
```

## 1-1000位置的拟合曲线图

```python
y_MB=[x[1] for x in list_MB]
x_MB=[x[0] for x in list_MB]
plt.figure(figsize=[16,8])
plt.xlabel('N')
plt.ylabel('D')
plt.xlim([-1,1000])
plt.plot(x_MB,y_MB)
plt.grid()
```
