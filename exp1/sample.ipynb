# 开始采样实验

##   随机采样函数

```python
# 随机采样函数
def random_sample(data,num1=50,num2=100,num3=500):
    # 随机采样100个点
    sample1=np.random.choice(data,num1)
    # 随机采样1000个点
    sample2=np.random.choice(data,num2)
    # 随机采样5000个点
    sample3=np.random.choice(data,num3)

    plt.figure(figsize=(10,10)) #设置窗口大小
    plt.suptitle('random sample result') # 图片名称

    plt.subplot(3,2,1)
    plt.hist(data)
    plt.title('origin')

    plt.subplot(3,2,2)
    plt.hist(sample1)
    plt.title('sample '+str(num1))

    plt.subplot(3,2,3)
    plt.hist(sample2)
    plt.title('sample '+str(num2))

    plt.subplot(3,2,4)
    plt.hist(sample3)
    plt.title('sample '+str(num3))
```

## 连续采样函数

```python
# 连续采样函数
def continuous_sample(data,num1=50,num2=100,num3=500):
    head=np.random.randint(1,5000)# 连续采样起始点(1-5000内随机，即我们在数据集中随机采样连续的一段)
    # 随机采样100个点
    ind1=np.arange(head,head+num1)
    sample1=data[ind1]

    # 随机采样1000个点
    ind2=np.arange(head,head+num2)
    sample2=data[ind2]

    # 随机采样5000个点
    ind3=np.arange(head,head+num3)
    sample3=data[ind3]

    plt.figure(figsize=(10,10)) #设置窗口大小
    plt.suptitle('continuous sample result') # 图片名称

    plt.subplot(3,2,1)
    plt.hist(data)
    plt.title('origin')

    plt.subplot(3,2,2)
    plt.hist(sample1)
    plt.title('sample '+str(num1))

    plt.subplot(3,2,3)
    plt.hist(sample2)
    plt.title('sample '+str(num2))

    plt.subplot(3,2,4)
    plt.hist(sample3)
    plt.title('sample '+str(num3))
```

##  等距采样函数

```python
# 等距采样函数
def Isometric_sample(data,distance=10,num1=50,num2=100,num3=500):
    head=0# 采样起始点
    # 随机采样100个点
    ind1=np.arange(head,head+num1,distance)
    sample1=data[ind1]

    # 随机采样1000个点
    ind2=np.arange(head,head+num2,distance)
    sample2=data[ind2]

    # 随机采样5000个点
    ind3=np.arange(head,head+num3,distance)
    sample3=data[ind3]

    plt.figure(figsize=(10,10)) #设置窗口大小
    plt.suptitle('isometric sample result') # 图片名称

    plt.subplot(3,2,1)
    plt.hist(data)
    plt.title('origin')

    plt.subplot(3,2,2)
    plt.hist(sample1)
    plt.title('sample '+str(num1))

    plt.subplot(3,2,3)
    plt.hist(sample2)
    plt.title('sample '+str(num2))

    plt.subplot(3,2,4)
    plt.hist(sample3)
    plt.title('sample '+str(num3))
```

# 测试中心极限定理

```python
test_=data[:,4]
plt.hist(test_)
plt.title('Exponential distribution function')
```

```python
# 从指数分布中随机采样500个样本点，得到样本点之和sum
# 重复1000次这样的采样，得到1000个新样本点，这些样本点服从正态分布

sample_list=[]
for i in range(1000):
    sample_=np.random.choice(test_,500)
    sum=sample_.sum()/500
    sample_list.append(sum)
plt.hist(sample_list)
plt.title('Central limit theorem')
```
